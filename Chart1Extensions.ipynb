```python
import requests
```

```python
countries = ["SSF"]
```

```python
data=[]
for country in countries:
  url='https://api.worldbank.org/v2/country/'+country+'/indicator/SH.STA.SUIC.FE.P5?format=json'
  json_data=requests.get(url).json()[1]
  data.append(json_data)
  print (url)
```

```python
import numpy as np
import pandas as pd
```

```python
flat_data=np.array(data).flatten()
```

```python
df=pd.json_normalize(flat_data)
df
df=df.dropna()
df
```

```python
df.to_csv('Projectchart1extension2.csv')
```
