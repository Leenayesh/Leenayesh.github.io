```python
import requests
import os
import json
from google.colab import files

```

```python
url_base = "https://api.stlouisfed.org/fred/series/observations?series_id={}&api_key=22ee7a76e736e32f54f5df0a7171538d&file_type=json"
file_base = "FRED_data_{}.json"
fredSeries = ['FPCPITOTLZGUSA','MICH','MEDCPIM158SFRBCLE','USACPICORMINMEI']
for i in fredSeries:
  URL = url_base.format(i)
  print(URL)
  data = requests.get(URL).json()
  print(data)
```

```python
fileName = file_base.format(i)
print(fileName)
```

```python
with open(fileName, 'w', encoding="utf-8") as f:
  json.dump(data, f, ensure_ascii=False,indent=4)
  files.download('{}'.format(fileName))
```
